```python
#2018
import pandas as pd

df = pd.read_csv("df_rg2018.csv")

df.rename(columns={"RADIACAO GLOBAL (KJ/mï¿½)": 'radiacao_global',
    "PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)": 'pressao_atmosferica_estacao',
    "PRESSï¿½O ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)": 'pressao_atmosferica_max',
    "PRESSï¿½O ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)": 'pressao_atmosferica_min',
    "TEMPERATURA DO AR - BULBO SECO, HORARIA (ï¿½C)": 'temperatura_ar_bulbo',
    "TEMPERATURA DO PONTO DE ORVALHO (ï¿½C)": 'temperatura_ponto_orvalho',
    "TEMPERATURA Mï¿½XIMA NA HORA ANT. (AUT) (ï¿½C)": 'temperatura_max',
    "TEMPERATURA Mï¿½NIMA NA HORA ANT. (AUT) (ï¿½C)": 'temperatura_min',
    "TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (ï¿½C)": 'temperatura_orvalho_max',
    "TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (ï¿½C)": 'temperatura_orvaho_min',
    "UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)": 'umidade_rel_max',
    "UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)": 'umidade_rel_min',
    "UMIDADE RELATIVA DO AR, HORARIA (%)": 'umidade_rel_ar',
    "VENTO, DIREï¿½ï¿½O HORARIA (gr) (ï¿½ (gr))": 'vento_direcao',
    "VENTO, RAJADA MAXIMA (m/s)": 'vento_rajada',
    "VENTO, VELOCIDADE HORARIA (m/s)": 'vento_velocidade'}, inplace=True)

df['data'] = pd.to_datetime(df['data'], format='%Y-%m-%d')
df['hora'] = pd.to_datetime(df['hora'], format='%H:%M').dt.hour

colunas_interesse = [
    'radiacao_global',
    'pressao_atmosferica_estacao',
    'pressao_atmosferica_max',
    'pressao_atmosferica_min',
    'temperatura_ar_bulbo',
    'temperatura_ponto_orvalho',
    'temperatura_max',
    'temperatura_min',
    'temperatura_orvalho_max',
    'temperatura_orvaho_min',
    'umidade_rel_max',
    'umidade_rel_min',
    'umidade_rel_ar',
    'vento_direcao',
    'vento_rajada',
    'vento_velocidade',
    'precipitacao'
]

for coluna in colunas_interesse:
    df[coluna] = df[coluna].astype(str).str.strip()
    df[coluna] = df[coluna].replace({'-9999': pd.NA, '-9999.0': pd.NA, '-9999,0': pd.NA, '-9999 ': pd.NA})
    df[coluna] = df[coluna].replace(-9999, pd.NA)

for coluna in colunas_interesse:
    df[coluna] = (
        df[coluna]
        .str.replace(',', '.', regex=False)
    )
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce')

df_rad_filtrado = df[(df['hora'] >= 9) & (df['hora'] <= 23)]
df_rad_media = (
    df_rad_filtrado.groupby(['data', 'cidade'])["radiacao_global"]
    .mean()
    .reset_index()
)

df_precipitacao_soma = (
    df.groupby(['regiao_intermediaria', 'cidade', 'data'])["precipitacao"]
    .sum()
    .reset_index()
)

df_outros_media = (
    df.groupby(['regiao_intermediaria', 'cidade', 'data'])[colunas_interesse[1:-1]]
    .mean()
    .reset_index()
)

df_media_diaria = pd.merge(df_precipitacao_soma, df_rad_media, on=['cidade', 'data'])
df_media_diaria = pd.merge(df_media_diaria, df_outros_media, on=['cidade', 'data', 'regiao_intermediaria'])

colunas_arredondar = [col for col in colunas_interesse if col != "precipitacao"]
df_media_diaria[colunas_arredondar] = df_media_diaria[colunas_arredondar].round(1)

df_media_diaria = df_media_diaria[['regiao_intermediaria', 'cidade', 'data'] + [col for col in df_media_diaria.columns if col not in ['regiao_intermediaria', 'cidade', 'data']]]
df_media_diaria = df_media_diaria.sort_values(by=['regiao_intermediaria', 'cidade', 'data'])
df_media_diaria = df_media_diaria.reset_index(drop=True)

df_media_diaria.to_csv('df_rs2018_dia.csv', index=False)
```
