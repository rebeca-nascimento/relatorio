```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

variaveis = [
    'precipitacao','radiacao_global', 'pressao_atmosferica_estacao', 'pressao_atmosferica_max',
    'pressao_atmosferica_min', 'temperatura_ar_bulbo', 'temperatura_ponto_orvalho',
    'temperatura_max', 'temperatura_min', 'temperatura_orvalho_max',
    'temperatura_orvaho_min', 'umidade_rel_max', 'umidade_rel_min',
    'umidade_rel_ar', 'vento_direcao', 'vento_rajada', 'vento_velocidade'
]


def gerar_matriz_correlacao(ano):
    df = pd.read_csv(f'df_semana{ano}.csv')

    matriz_correlacao = df[variaveis].corr()

    plt.figure(figsize=(12, 8))
    sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

    plt.title(f'Matriz de Correlação - {ano}')
    plt.savefig(f'matriz_correlacao_{ano}.png', bbox_inches='tight')
    plt.show()

for ano in [2018, 2019, 2020, 2021, 2022, 2023]:
    gerar_matriz_correlacao(ano)
```

```python
import pandas as pd
df2018 = pd.read_csv('df_semana2018.csv')
df2019 = pd.read_csv('df_semana2019.csv')
df2020 = pd.read_csv('df_semana2020.csv')
df2021 = pd.read_csv('df_semana2021.csv')
df2022 = pd.read_csv('df_semana2022.csv')
df2023 = pd.read_csv('df_semana2023.csv')

df_total = pd.concat([df2018, df2019, df2020, df2021, df2022, df2023], ignore_index=True)

df_total.to_csv('df_total.csv', index=False)
```
